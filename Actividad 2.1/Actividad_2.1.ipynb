```python
#Convertir archivo filtrado a CSV
#Variables filtradas:Nos filtro la primera fila con sus respectivas columnas   
filtro5=df.head()
filtro5.to_csv("Filtro5.csv")
```

```python
#Convertir archivo filtrado a CSV
#variables filtradas:Nos filtro la ultima fila con sus respetivas columnas 
filtro6=df.tail()
filtro6.to_csv("Filtro6.csv")
```

```python
#Convertir archivo filtrado a CSV
#variables filtradas:Lo filtre por la columna CANT fuera mayor a 7
filtro7=df[df["CANT"] > 7]
filtro7.to_csv("Filtro7.csv")
```

```python
#Convertir archivo filtrado a CSV
#Variables filtradas:Lo filtre por que el PRECIO_UNITARIO fuera mayor a 1000 y el NOMBRE_VENDEDOR fuera ALFREDO CANELA
filtro8=df[(df["PRECIO_UNITARIO"] < 1000) & (df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro8.to_csv("Filtro8.csv")
```

```python
#Convertir archivo filtrado a CSV
#Variables filtradas:Lo filtre porque el PRECIO_UNITARIO fuera mayor a 1000 o que el NOMBRE_VENDEDOR fuera ALFREDO CANELA
filtro9=df[(df["PRECIO_UNITARIO"] < 1000) | (df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro9.to_csv("Filtro9.csv")
```

```python
#Convertir archivo filtrado a CSV
#Variables filtradas:Lo filtre por que en NOMBRE_CIENTE no fuera ELICAMEX
filtro10=df[~(df["NOMBRE_CLIENTE"] == "ELICAMEX")]
filtro10.to_csv("Filtro10.csv")
```
